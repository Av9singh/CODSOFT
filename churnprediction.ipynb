```python
import pandas as pd
from sklearn.utils import resample

# Assuming 'churn' DataFrame is defined and contains the dataset

# Separate majority and minority classes
churn_majority = churn[churn['exited'] == 0]
churn_minority = churn[churn['exited'] == 1]

# Downsample majority class
churn_majority_downsampled = resample(churn_majority, n_samples=2037, replace=False, random_state=42)

# Concatenate minority class and downsampled majority class
churn_df = pd.concat([churn_majority_downsampled, churn_minority])

# Display value counts of the 'exited' column
print(churn_df['exited'].value_counts().to_frame())
```

```python
x=churn_df.drop(['exited'],axis=1)
y=churn_df['exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=500)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score,accuracy_score,f1_score,ConfusionMatrixDisplay
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
cmd=ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,y_pred,labels=lr.classes_),display_labels=lr.classes_)
cmd.plot()
```
